### Replacing the null values 

```python
A["Ozone"]=A["Ozone"].fillna(A["Ozone"].mean())
A["Solar.R"]=A["Solar.R"].fillna(A["Solar.R"].mean())
```

```python
A
```

```python
A.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
A
```

### 1.Pie chart

```python
plt.pie
```

```python
labels=['Ozone','Solar.R','Wind','Temp']
```

```python
A.columns
```

```python
sizes=[A['Ozone'].mean(),A['Solar.R'].mean(),A['Wind'].mean(),A['Temp'].mean()]
```

```python
colors=['red','green','pink','yellow']
```

```python
explode=(0.1,0,0,0)
```

```python
plt.pie(sizes,explode=explode,labeldistance=1.1,labels=labels,
colors=colors,autopct='%0.1f%%')
plt.title('Average Data')
# plt.figure(figsize=(5,5))
plt.show()
```

### 2.Bar plot

```python
def addlabels():
    for i in range(len(A.iloc[0:21,3])):
        plt.text(i,A.iloc[0:21,3][i],A.iloc[0:21,3][i],ha='center')
                 
y=np.arange(len(A.iloc[0:21,3]))
plt.figure(figsize=(10,6))
addlabels()
plt.bar(x=y,height=A.iloc[0:21,3],tick_label=y,color="red")
plt.title('Wind measure for 1st-20th May')
plt.xlabel('Days')
plt.ylabel('Wind')
plt.show()
```

### 3.Box_plot

```python
A.boxplot(column=['Ozone','Wind','Temp','Solar.R'],grid=False)
plt.title('Box-Plot')
plt.suptitle('Temperature')
plt.ylabel('range')
plt.savefig('plot3.png')
```

### 4.Histogram

```python
plt.figure(figsize=(10,6))
plt.hist(x=A['Temp'],bins=20,color="pink",label="Temp")
plt.title("Temperature count")
plt.xlabel("Temperature")
plt.ylabel("count")
plt.plot()
```

### 5.Scatter Plot

```python
plt.figure(figsize=(10,6))
plt.scatter(A.iloc[0:30,5],A.iloc[0:30,3],color="green")
plt.title("Wind measures of May month")
plt.xlabel("Day")
plt.ylabel("Wind")
```

### 6.Pair Plot

```python
import seaborn

seaborn.pairplot(A.iloc[:,[0,1,2,3,6]],hue='humidity')
plt.show()
```

### 7.Heatmap

```python
seaborn.heatmap(data=A.iloc[0:15,[0,1,2,3]],annot=True,linewidths=5,linecolor='yellow',square=True)
plt.show()
```

```python
seaborn.heatmap(A.corr(),annot=True,square=True)
```

### 8.Line graph

```python
plt.figure(figsize=(10,6))
plt.plot(A.iloc[0:30,5],A.iloc[0:30,3],color="blue")
plt.title('Temperature of May month')
plt.xlabel("Day")
plt.ylabel('Temperature')
```
